# Matplotlib and Seaborn

Matplotlib and Seaborn are python libraries taht are used for data visualization. They have inbuilt modules for plotting different graphs. While Matplotlib is used to embed graphs into applications, Seaborn is primarily used for statistical graphs.

But when should we use either of the two? Let's understand this with the help of a comparative analysis. Below are the comparison between  Python's two well-known visualization packages Matplotlib and Seaborn.

## Matplotlib
- It is used for basic graph plotting like line charts, bar graphs, etc.
- It mainly works with datasets and arrays.
- Matplotlib acts productively with data arrays and frames. It regards the aces and figures as objects.
- Matplotlib is more customizable and pairs well with Pandas and Numpy for Exploratory Data Analysis.

## Seaborn
- It is mainly used for statistics visualization and can perform complex visualizations with fewer commands.
- It works with entire datasets.
- Seaborn is considerably more organized and fuctional that Matplotlib and treats the entire dataset as a solitary unit.
- Seaborn has more inbuilt themes and is mainly used for statistical analysis.

## Line Chart

A Line Chart is a graph that represents information as a series of data points connected by a straight line. In line charts, each data point or marker is plotted and connected with a line or curve.

Let's consider the apple yield (tons per hectare) in Brisbane. Let's plot a line graph using this data and see how the yield of apples changes over time. We are going to both by using Matplotlib and Seaborn as follows :

### Using Matplotlib

We are using random  data points to represent the yield of apples.

```python
yield_apples = [0.89, 0.91, 0.92, 0.93, 0.94, 0.95] #ton per hectare
```

After we define the data points of apple yields, we are going to plot its Line Chart using Matplotlib.

```python
plt.plot(yield_apples)
```

To better understand the graph and its purpose, we can add the x-axis values too.

```python
years = [2017, 2018, 2019, 2020, 2021, 2022]
yield_apples = [0.89, 0.91, 0.92, 0.93, 0.94, 0.95]
```

Since we have the 'years' variable as the x-axis, we need to type it before the yield_apples variable below.

```python
plt.plot(years, yield_apples) #plt.plot(x, y)
```

Let's add labels to the axes so that we can show what each axis represents.

```python
plt.plot(years, yield_apples)
plt.xlabel('Year')
plt.ylabel('Yield 9tons per hectare)');
```

To plot multiple datasets on the same graph, just use the plt.plot function once for each dataset. Let's use this to compare the yields of apples vs. oranges on the same graph.

```python
apples = [0.89, 0.91, 0.93, 0.97, 0.986, 0.943]
tangerines = [0.79, 0.61, 0.93, 0.97, 0.962, 0.931]
```

```python
plt.plot(years, apples)
plt.plot(years, tangerines)
plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)');
```

We can add a legend which tells us what each line in our graph means. To understand what we are plotting, we can add a title to our graph.

```python
plt.plot(years, apples)
plt.plot(years, tangerines)

plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')

plt.title("Crop Yields in Brisbane")
plt.legend(['Apples', 'Tangerines'])
```

To show each data point on our graph, we can highlight them with markers using the marker argument. Many different marker shapes like a circle, cross, square, diamond, etc. are provided by Matplotlib.

```python
plt.plot(years, apples, marker='o')
plt.plot(years, tangerines, marker='x')

plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')

plt.title("Crop Yields in Brisbane")
plt.legend(['Apples', 'Tangerines'])
```

You can use the plt.figure function to change the size of the figure.

```python
plt.figure(figsize=(12,6))

plt.plot(years, tangerines, marker='o')
plt.title("Yield of Tangerines (tons per hectare)")
```

## Bar Graph

When you have categorical data, you can represent it with a bar graph. A bar graph plots data with the help of bars, which represent value on the y-axis and category on the x-axis. Bar graphs use bars with varying heights to show the data which belongs to a specific category.

### Using Matplotlib

```python
plt.bar(years, tangerines) #plt.bar(x, y)

plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare')

plt.title("Crop Yields in Brisbane")
```

```python
plt.bar(years, apples)
plt.bar(years, tangerines, bottom=apples)
```

### Using Seaborn

Let's use the data tips dataset in Seaborn next. The dataset consists of :

1. Information about the sex gender)
2. Time of day
3. Total bill
4. Tips given by customers visiting the restaurant for a week

This 'Tips Dataset' is a dataset provided by seaborn library, so you can directly load the dataset.

```python
tips_df = sns.load_dataset("tips")
tips_df.head()
```

We can draw a bar chart to visualize how the average bill amount varies across different days of the week. We can do this by computing the day-wise averages and than using plt.bar. The Seaborn library also provides a barplot function that automatically compute averages.

```python
sns.barplot(x='day',
            y='total_bill',
            data=tips_df)
```

If you want to compare bar plots side-by-side, you can use the hue argument. The comparison will be done based on the third feature specified in this argument.

```python
sns.barplot(x='day', y='total_bill', hue='sex', data = tips_df)
```

You can make the bars horizontal by switching the axes.

```python
sns.barplot(x='total_bill', y='day', hue='sex', data = tips_df)
```

# Histograms

A Histogram is a bar representation of data that varies over a range. It plots the height the data belonging to a range along the y-axis and the range along the x-axis. Histograms are used to plot data over a range of values. They use a bar representation to show the data belonging to each range. Let's again use the data 'Iris' dataset which contains information about flowers to plot histograms.

The 'Iris' dataset is a dataset provided by Seaborn which is **completely numerical**.

```python
iris_df = sns.load_dataset("iris")
iris_df.head(3)
```

### Using Matplotib

We are going to plot a histogram from the 'Sepal Width' column.

```python
iris_df.sepal_width
```

```python
plt.title("Distribution of Sepal Width")
plt.hist(iris_df.sepal_width);
```

We can control the number or size of bins too.

```python
# specifying the number of bins
plt.hist(iris_df.sepal_width, bins=5)
```

We can change the number and size of bins using numpy too.

```python
import numpy as np

#specifying the number of bins
plt.hist(iris_df.sepal_width, bins=np.arange(2, 5, 0.25));
```

We can create bins of unequal size too.

```python
# bins of unequal sizes
plt.hist(iris_df.sepal_width, bins=[1,3,4,4,5])
```
